```python
import numpy as np
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python

df=pd.read_csv('drive/MyDrive/Colab Notebooks/concrete_data.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
columns = df.columns
```

```python
X = df[columns[columns != "Strength"]]
X.head()
```

```python
y = df["Strength"]
y.head()
```

# Modeling Process
---
Step 1: Randomly split the data into training and test sets

```python



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Step 2: Build the model

```python

model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(1))  # Output layer (assuming regression problem)
```

Compile the model

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

Step 3: Train the model

```python
history = model.fit(X_train, y_train, epochs=50, verbose=0)


```

 Step 4: Evaluate the model 50 times and compute mean squared errors

```python
num_repeats = 50
mse_list = []

for _ in range(num_repeats):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=np.random.randint(1, 100))
    model.fit(X_train, y_train, epochs=50, verbose=0)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

```

Step 5: Report mean and standard deviation of mean squared errors

```python

mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

print("Mean of Mean Squared Errors:", mean_mse)
print("Standard Deviation of Mean Squared Errors:", std_mse)
```
